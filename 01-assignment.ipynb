```python
import numpy as np
import matplotlib.pyplot as plt

def buat_sinyal_sinus(amplitudo, frekuensi, durasi):

    sampling_rate = 100
    t = np.linspace(0, durasi, int(durasi * sampling_rate))

    sinyal = amplitudo * np.sin(2 * np.pi * frekuensi * t)

    return sinyal, t

amplitudo = 2
frekuensi = 5
durasi = 2

sinyal, waktu = buat_sinyal_sinus(amplitudo, frekuensi, durasi)

plt.plot(waktu, sinyal)
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.title('Sinyal Sinusoidal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace (0, 2, 100)

sinyal1 = np.sin(2 * np.pi * 1 * t )

amp = 2
sinyal1_amp = sinyal1 * amp

flg, (ax1, ax2) = plt.subplots(2, 1, figsize = (7, 8))

ax1.plot(t, sinyal1)
ax1.set_title('Sinyal Asli')
ax1.set_ylabel('Amplitudo')
ax1.grid(True)

ax2.plot(t, sinyal1_amp)
ax2.set_title('Penguatan Sinyal')
ax2.set_ylabel('Amplitudo')
ax2.grid(True)


plt.tight_layout()
plt.show()
```
