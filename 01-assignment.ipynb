```python
import numpy as np
import matplotlib.pyplot as plt

waktu = np.linspace(0, 2, 1000)

sinyal_asli = np.sin(2 * np.pi * 5 * waktu)

amplitudo = 2
sinyal_dikuatkan = sinyal_asli * amplitudo

fig, (grafik1, grafik2) = plt.subplots(2, 1, figsize=(7, 8))

grafik1.plot(waktu, sinyal_asli)
grafik1.set_title('Sinyal Asli')
grafik1.set_ylabel('Amplitudo')
grafik1.grid(True)

grafik2.plot(waktu, sinyal_dikuatkan)
grafik2.set_title('Sinyal yang Dikuatkan')
grafik2.set_ylabel('Amplitudo')
grafik2.grid(True)

plt.tight_layout()
plt.show()
```
