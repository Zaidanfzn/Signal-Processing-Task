```
import numpy as np
import matplotlib.pyplot as plt

def generate_sinusoidal_signal(amplitude, frequency, duration):
  """
  Membangkitkan sinyal sinusoidal.

  Args:
    amplitude: Amplitudo sinyal.
    frequency: Frekuensi sinyal dalam Hz.
    duration: Durasi sinyal dalam detik.

  Returns:
    Array NumPy yang berisi nilai-nilai sinyal sinusoidal.
  """

  # Waktu sampling (sesuaikan sesuai kebutuhan)
  sampling_rate = 1000  # 1000 sampel per detik
  time = np.linspace(0, duration, int(duration * sampling_rate))

  # Bangkitkan sinyal sinus
  signal = amplitude * np.sin(2 * np.pi * frequency * time)

  return signal, time

# Atur parameter sinyal
amplitude = 2  # Atur amplitudo sesuai keinginan
frequency = 5  # Atur frekuensi sesuai keinginan
duration = 2  # Atur durasi sinyal dalam detik

# Bangkitkan sinyal
sinyal, waktu = generate_sinusoidal_signal(amplitude, frequency, duration)

# Plot sinyal
plt.plot(waktu, sinyal)
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.title('Sinyal Sinusoidal')
plt.grid(True)
plt.show()
```
